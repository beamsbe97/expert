```python
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_profile = pd.read_excel("A2 - Profile.xlsx")
df_survey = pd.read_excel("A2 - Survey.xlsx")
```

```python
df_merged = df_profile.merge(df_survey, left_on="CustomerID", right_on="customerID", how="inner")
df_merged
```

```python
## preprocessing
enc = OrdinalEncoder()

#hot encoding cetegorical features
df_merged[["Job", "Marital_status", "Education", "Credict_Payment", "Shopping type", "LoyaltyCard_of_Others?"]] = enc.fit_transform(df_merged[["Job", "Marital_status", "Education", "Credict_Payment", "Shopping type", "LoyaltyCard_of_Others?"]])


#drop blank rows
df_merged = df_merged.dropna(subset=["Marital_status", "Credict_Payment", "LoyaltyCard_of_Others?", "Product-quality", "Deli-quality", "customerID", "Service-quality", "Shopping_comfort"])
df_merged = df_merged.drop(columns=["CustomerID", "customerID"])
```

```python
df_merged
```

```python

#split into training set and test set
df_train = df_merged[:1790]
df_test = df_merged[1790:]
```

```python
#fit
X_train = df_train[["Age", "Job", "Marital_status", "Education", "Credict_Payment", "Product-quality", "Deli-quality", "Service-quality", "Shopping_comfort"]]
Y_train = df_train["LoyaltyCard_of_Others?"]
clf = DecisionTreeClassifier(min_samples_split=45)
clf = clf.fit(X_train, Y_train)
```

```python
#test
X_test = df_test[["Age", "Job", "Marital_status", "Education", "Credict_Payment", "Product-quality", "Deli-quality", "Service-quality", "Shopping_comfort"]]
Y_test = df_test["LoyaltyCard_of_Others?"]

predicted = clf.predict(X_test)
accuracy = accuracy_score(predicted, Y_test)
print(accuracy)
```

```python
fig, ax = plt.subplots(figsize=(27, 17)) 
plot_tree(clf, ax=ax, fontsize=8)
plt.show()
```
